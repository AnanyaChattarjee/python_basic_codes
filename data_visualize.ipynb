```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

importing libraries

```python
df = sns.load_dataset('titanic')
```

loading dataset into df 

```python

print(df.info())
print(df.head())
```

```python
df.drop(['deck', 'embark_town', 'alive', 'who', 'adult_male', 'class'], axis=1, inplace=True)
```

deleting useless rows

```python
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

filling NaN values

```python
df.dropna(inplace=True)
```

droping NaN values

```python
df
```

```python
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df['alone'] = df['alone'].astype(int) 
```

performing label encoding

```python

sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()

```

bar chart of survival count

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

bar chart of survival count by gender

```python
sns.countplot(x='survived', hue='pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

survival count by passenger class

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

age distribution of passengers on titanic

```python
sns.histplot(df['fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title("Fare vs Age by Survival")
plt.show()
```

fare vs age by survival scatter plot
