TRANSACTIONS ARE PRETTY HIGH IN DECEMBER AND QUITE LOW IN JANUARY

```python
Q1 = data['amount (INR)'].quantile(0.25)
Q3 = data['amount (INR)'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data['amount (INR)'] < (Q1 - 1.5 * IQR)) | (data['amount (INR)'] > (Q3 + 1.5 * IQR))]
print(f"Number of outliers in 'amount': {len(outliers)}")
```

```python
threshold = data['amount (INR)'].quantile(0.99)
high_value = data[data['amount (INR)'] >= threshold]

print(f"High-value transactions (top 1%): {len(high_value)}")
display(high_value[['timestamp', 'amount (INR)']].sort_values(by='amount (INR)', ascending=False).head())
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], errors='coerce')

data['day'] = data['timestamp'].dt.date
```

```python
daily_avg = data.groupby('day')['amount (INR)'].mean()

rolling_avg = daily_avg.rolling(window=7).mean()

plt.figure(figsize=(12, 5))
plt.plot(daily_avg.index, daily_avg, label='Daily Avg', alpha=0.5)
plt.plot(rolling_avg.index, rolling_avg, label='7-Day Rolling Avg', color='red')
plt.title("Daily vs Smoothed Transaction Amount")
plt.xlabel("Date")
plt.ylabel("Avg Transaction Amount ")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = data.select_dtypes(include='number').dropna()
scaler = StandardScaler()
scaled = scaler.fit_transform(features)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled)

plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.3, c='blue')
plt.title("PCA: Transaction Pattern Clusters")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```
